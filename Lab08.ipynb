# Pregunta 1

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img1 = cv2.imread('Imagenes/1.jpeg')
img1_fix = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1_fix)
```

```python
img2 = cv2.imread('Imagenes/2.jpeg')
img2_fix = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2_fix)
```

```python
img3 = cv2.imread('Imagenes/3.jpg')
img3_fix = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
plt.imshow(img3_fix)
```

```python
alto1, ancho1 = img1.shape[:2]
alto2, ancho2 = img2.shape[:2]
alto3, ancho3 = img3.shape[:2]
alto = max(alto1,alto2,alto3)
ancho = max(ancho1,ancho2,ancho3)
print(alto , ancho)
```

```python
img1_fix = cv2.resize(img1_fix,(ancho,alto))
plt.imshow(img1_fix)
```

```python
img2_fix = cv2.resize(img2_fix,(ancho,alto))
plt.imshow(img2_fix)
```

```python
img3_fix = cv2.resize(img3_fix,(ancho,alto))
plt.imshow(img3_fix)
```

# Pregunta 2

```python
img1_red = img1_fix.copy()
img1_red[:,:,1] = 0
img1_red[:,:,2] = 0
plt.imshow(img1_red)
```

```python
img2_green = img2_fix.copy()
img2_green[:,:,0] = 0
img2_green[:,:,2] = 0
plt.imshow(img2_green)
```

```python
img3_blue = img3_fix.copy()
img3_blue[:,:,0] = 0
img3_blue[:,:,1] = 0
plt.imshow(img3_blue)
```

```python
new_img = img1_red + img2_green + img3_blue
plt.imshow(new_img)
```

# Pregunta 3

```python
negativo = 255 - new_img
plt.imshow(negativo)
```

```python
cv2.imwrite('Imagenes/negativo.jpg', negativo)
```

```python
gris = cv2.cvtColor(negativo, cv2.COLOR_BGR2GRAY)
plt.imshow(gris)
```

```python
cv2.imwrite('Imagenes/negativo_gray.jpg', gris)
```

# Pregunta 5

```python
# Utilizamos la imagen 3 como base
img_descripcion = img3.copy()

# circulo
centro = (470, 360)
radio = 210
color_circulo = (0, 255, 0)# Cirgulo color verde
cv2.circle(img_descripcion, centro, radio, color_circulo, thickness=3)

# Texto
texto = "Perro"
cv2.putText(img_descripcion, texto, (centro[0] - 40, centro[1] - radio - 10), 
            cv2.FONT_HERSHEY_SIMPLEX, 2, (255, 255, 255), thickness=2, lineType=cv2.LINE_AA)

# Mostramos imagen
cv2.imshow('Figura Anotada', img_descripcion)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Guardamos imagen
cv2.imwrite('Imagenes/img_descricion.jpg', img_descripcion)
```
