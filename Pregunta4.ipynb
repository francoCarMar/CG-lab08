```python
import matplotlib.pyplot as plt
import cv2

img = cv2.imread('Imagenes/3.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Menu
print("Bienvenido. Elija una opción para alternar la visibilidad de los canales de color:")
print("0 - Mostrar solo canal Rojo")
print("1 - Mostrar solo canal Verde")
print("2 - Mostrar solo canal Azul")
print("3 - Mostrar imagen completa")
print("4 - Salir")

while True:
    # validacion de entrada
    try:
        option = int(input("Opción (0-4): "))
    except ValueError:
        print("Entrada no válida. Intente de nuevo.")
        continue

    if option == 4:
        print("Hasta luego.")
        break

    temporal = img.copy()
    
    # dejamos solo el canalar correspondiente
    if option == 0: # rojo
        temporal[:, :, 1] = 0
        temporal[:, :, 2] = 0
    elif option == 1: # verde
        temporal[:, :, 0] = 0
        temporal[:, :, 2] = 0
    elif option == 2: # azul
        temporal[:, :, 0] = 0
        temporal[:, :, 1] = 0
    elif option == 3:
        pass
    else:
        print("Opción no válida. Intente con 0-4.")
        continue

    # Mostramos imagen correspondiente
    plt.imshow(temporal)
    plt.axis('off')
    plt.title(f"Canal mostrado: {['Rojo', 'Verde', 'Azul', 'Completo'][option]}")
    plt.show()

```
